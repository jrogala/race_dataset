```python
import fastf1
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
race = fastf1.get_session(2022, 'Bahrain Grand Prix', 'R')
race.load()
```

```python
from matplotlib import pyplot as plt
df = pd.DataFrame(race.car_data["16"]).set_index("Date")
plt.figure(figsize=(15, 15))
plt.plot(df.loc[:, ["DRS", "Speed"]].dropna().groupby("DRS").max())
plt.show()
```
